```python
import pandas as pd
import matplotlib.pyplot as plt
test = pd.read_excel('1.xlsx')
```

```python
netflix = pd.read_excel('1.xlsx',sheet_name=0)
amazon = pd.read_excel('1.xlsx',sheet_name=1)
```

```python
netflix["service"] = "netflix"
amazon["service"] = "amazon"
```

```python
netflix
```

```python
df = pd.concat([netflix,amazon])
df
```

```python
graph = df[['service','type','show_id']].groupby(['service','type']).count()
```

```python
# make a histogram of the data frame
graph.unstack().plot(kind='bar',stacked=True,figsize=(10,10))
plt.show()
# calculate the ratio of tv shows to total content, and put the number into the bars
graph = graph.reset_index()
graph['ratio'] = graph['show_id']/graph['show_id'].sum()
graph
```

```python
data1 = netflix[['release_year', 'show_id']].groupby(['release_year']).count()
data2 = amazon[['release_year', 'show_id']].groupby(['release_year']).count()
```

```python
data1
```

```python
graph1 = data1.plot(kind='bar',figsize=(10,10))
```

```python
graph2 = data2.plot(kind='bar',figsize=(10,10))
```
